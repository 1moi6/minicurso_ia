```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados

n_points = 1000
X_data = np.random.uniform(-1.5, 1.5, (n_points, 2))
radii = np.linalg.norm(X_data, axis=1, ord=2)

def sigmoid(x): return 1 / (1 + np.exp(-x))
sigma_r = .05
probs = sigmoid((1 - radii) / sigma_r)
Y_data = np.random.binomial(1, probs).astype(np.float32)

# Plot dos pontos com rótulos e eixos centrais
plt.figure(figsize=(6, 6))
plt.scatter(X_data[:, 0], X_data[:, 1], c=Y_data, cmap='bwr', edgecolor='k', alpha=0.7)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)
plt.grid(True)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
