```python
import matplotlib.pyplot as plt
import numpy as np

# Define the input range
x = np.linspace(-10, 10, 500)

# Define activation functions
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
leaky_relu = np.where(x > 0, x, 0.01 * x)
softmax_input = np.vstack([x, np.ones_like(x)*-2, np.ones_like(x)*2])  # 3-class example
softmax = np.exp(softmax_input) / np.sum(np.exp(softmax_input), axis=0)
# Novo gráfico combinando Sigmoid, ReLU e Leaky ReLU

fig, ax = plt.subplots(figsize=(8, 6))

# Plot das funções
ax.plot(x, sigmoid, label='Sigmoide', linewidth=2)
ax.plot(x, relu, label='ReLU', linewidth=2)
ax.plot(x, leaky_relu, label='Leaky ReLU', linewidth=2)

# Configurações visuais
ax.set_title('Funções de Ativação: Sigmoide, ReLU e Leaky ReLU', fontsize=14)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True)
ax.legend()

# Mostrar todas as bordas da caixa (inclusive direita e superior)
for spine in ax.spines.values():
    spine.set_visible(True)

plt.tight_layout()
plt.show()
```
